```python
import numpy as np
import matplotlib.pyplot as plt

xpoints = np.array([10,35,6,90,50,51,60])
ypoints = np.array([20,66,70,35,66,99,70])

plt.plot(xpoints,ypoints, color = "g",linestyle = "dotted",marker = "o")

plt.title("SHARE MARKET")
plt.xlabel("INVESTORS")
plt.ylabel("INVESTINGS")

plt.show()
```
